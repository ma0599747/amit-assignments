```python
from sklearn.preprocessing import OneHotEncoder  

#creating instance of one-hot-encoder
encoder = OneHotEncoder()

#perform one-hot encoding on 'team' column 
encode_column = pd.DataFrame(encoder.fit_transform(data[["name","host_name","neighbourhood_group","neighbourhood","room_type"]]).toarray())
encode_column
```

```python
data.drop(['name','host_name','neighbourhood_group','neighbourhood',"room_type"],axis=1,inplace=True)
```

```python
x = new_df.drop('availability_365',axis=1)
y = data['availability_365']

x
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense
```

```python
model = keras.models.Sequential([
                                 

                                
                                 
                                 keras.layers.Dense(12, activation='relu',input_dim=10547 ),

                                 keras.layers.Dense(6, activation='relu'),

                                 keras.layers.Dense(3, activation='relu'),


                                 keras.layers.Dense(1, activation='relu')


                                 ])
```

```python
model.summary()
```

```python
callback_early_stopping=keras.callbacks.EarlyStopping(monitor="loss",patience=5)

#opt = keras.optimizers.Adam(learning_rate=0.01)
#model.compile(loss='categorical_crossentropy', optimizer=opt)


lr_schedule = keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-2,
    decay_steps=10000,
    decay_rate=0.9)
optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)




model.compile(optimizer="adam",loss="mean_squared_error",metrics=['mean_squared_error'])

# model.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
```

```python
model.fit(x,y, epochs = 50, validation_split= 0.1, callbacks = callback_early_stopping)
```
