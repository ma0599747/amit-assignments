```python
dataset.drop(["sex","race"],axis=1,inplace=True)


# dataset.columns
```

```python
from sklearn.preprocessing import OneHotEncoder  

#creating instance of one-hot-encoder
encoder = OneHotEncoder()

#perform one-hot encoding on 'team' column 
encode_column = pd.DataFrame(encoder.fit_transform(dataset[['workclass','education','marital-status','occupation',"relationship","native-country"]]).toarray())
encode_column
```

```python
dataset.dtypes
```

```python
dataset

new_df = pd.concat([dataset,encode_column],axis = 1)
new_df
```

```python
new_df.drop(['workclass','education','marital-status','occupation',"relationship","native-country"],axis=1,inplace=True)
```

```python
new_df.iloc[:,:-1].shape
```

```python
new_df
```

```python


from sklearn import preprocessing

le = preprocessing.LabelEncoder()

new_df= new_df.apply(le.fit_transform)
```

```python
# x=new_df.iloc[:,:-1]
# y=new_df.iloc[:,-1]

x = new_df.drop('income',axis=1)
y = dataset['income']
```

```python
y = le.fit_transform(y)
```

```python
model=keras.models.Sequential([
                               
                               
                               

                               keras.layers.Dense(8,activation="relu",input_dim=101),

                               keras.layers.BatchNormalization(),

                               keras.layers.Dropout(0.3),

                               keras.layers.Dense(4,activation="relu"),

                               keras.layers.BatchNormalization(),

                               keras.layers.Dense(1,activation="sigmoid")
                               
                              
])
```

```python
model.summary()
```

```python
x
```

```python
callback_early_stopping=keras.callbacks.EarlyStopping(monitor="loss",patience=5)

#opt = keras.optimizers.Adam(learning_rate=0.01)
#model.compile(loss='categorical_crossentropy', optimizer=opt)


# lr_schedule = keras.optimizers.schedules.ExponentialDecay(
#     initial_learning_rate=1e-2,
#     decay_steps=10000,
#     decay_rate=0.9)
# optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)




model.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")

# model.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
```

```python
model.fit(x,y, epochs = 50, validation_split= 0.1, callbacks = callback_early_stopping)
```
